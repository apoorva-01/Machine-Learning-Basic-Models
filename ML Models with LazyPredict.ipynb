# LazyClassifier

```python
# import necessary modules
import warnings
warnings.filterwarnings('ignore')
import time 
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from lazypredict.Supervised import LazyClassifier
```

```python
# load the iris dataset
data=load_iris()
X=data.data
y=data.target
```

```python
data
```

```python
# split the dataset 
X_train, X_test, y_train, y_test =train_test_split(X,y,test_size=.3,random_state =23)
classi=LazyClassifier(verbose=0,predictions=True)
```

```python
# fit and train the model 
start_time_1=time.time()
models_c,predictions_c=classi.fit(X_train, X_test, y_train, y_test)
end_time_1=time.time()
```

```python
models_c
```

# LazyRegression

```python
# import necessary modules
import warnings
warnings.filterwarnings('ignore')
import time 
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from lazypredict.Supervised import LazyRegressor
```

```python
df=fetch_california_housing()
```

```python
X1=df.data
y1=df.target
```

```python
X_train1,X_test1,y_train1,y_test1=train_test_split(X1,y1,random_state=134,test_size=.3)
regr=LazyRegressor(verbose=0,predictions=True)
```

```python
# fit and train the model 
start_time_2=time.time()
models_r,predictions_r=regr.fit(X_train1, X_test1, y_train1, y_test1)
end_time_2=time.time()
```

```python
models_r
```

```python
predictions_r
```

```python
# time complexity 
print("The time taken by LazyClassifier for {0} samples is {1} ms".format(len(data.data),round(end_time_1-start_time_1,0)))
print("The time taken by LazyRegressor for {0} samples is {1} ms".format(len(df.data),round(end_time_2-start_time_2,0)))
```
